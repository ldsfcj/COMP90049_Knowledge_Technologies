```python
def exist_Pre_Component(prefix,dictionalwords,distance1):
    for dic in dictionalwords:
        if (editdistance.eval(prefix,dic)<distance1):
#             print(dic)
            return True
    return False
    
def exist_Tail_Component(suffix,dictionalwords,distance2):
    for dic in dictionalwords:
        if (editdistance.eval(suffix,dic)<distance2):
#             print(dic)
            return True
    return False

distance1=3.65
distance2=2.31
expc=False
estc=False
for w in candidatewords:
    prefix = w[:2]
    suffix = w[2:]
    expc = exist_Pre_Component(prefix,dictionalwords,distance1)
    estc = exist_Tail_Component(suffix,dictionalwords,distance2)
    if(expc and estc):
        predict_blend.append(w)
        count += 1
        print(w)
print(count)
```
